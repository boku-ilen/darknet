```python
import pandas as pd
from collections import Counter
```

```python
# load data exported with ImagesDifference.ipynb
# data is sorted by modification time (dos)
df = pd.read_csv('ImagesDifference/diff_pixels.csv')
df
```

```python
df.rename(columns={'Unnamed: 0':'idx'}, inplace=True)
df['user'] = 0
df
```

```python
# take only rows with some differences
df_all_movement = df[df['diff pixels'] != 0]
df_all_movement
```

```python
# take only rows with strong differences (only few pixels are mainly rotating wind turbines but no other movement)
df_strong_movement = df_all_movement[df_all_movement['diff pixels'] > 2000]
# set new ids
df_strong_movement.insert(0, 'ID', range(0, len(df_strong_movement)))
df_strong_movement.set_index('ID', inplace=True)
df_strong_movement
df_strong_movement
```

```python
# group data into potential users
user = 1
for row in range(len(df_strong_movement)-1):
    if df_strong_movement.loc[row+1, 'list idx'] - df_strong_movement.loc[row, 'list idx'] > 2:
        df_strong_movement.loc[row, 'user'] = user
        user += 1
        df_strong_movement.loc[row+1, 'user'] = user
    else:
        df_strong_movement.loc[row, 'user'] = user
        df_strong_movement.loc[row+1, 'user'] = user
```

```python
users = Counter(df_strong_movement['user'])
#users
```

```python
# take data with minimum 30 rows (~30 seconds)
min_30_seconds_users = []
for user in range(len(users)):
    if users[user+1] > 30:
        min_30_seconds_users.append(user+1)
```

```python
# if images have different sizes, it must be different users or even different day 
df_min_30_seconds_strong_movement = df_strong_movement[df_strong_movement['user'].isin(min_30_seconds_users)]
print(Counter(df_min_30_seconds_strong_movement['same size']))
print(df_min_30_seconds_strong_movement[df_min_30_seconds_strong_movement['same size']==False])
```

```python
min_30_sec_users = Counter(df_min_30_seconds_strong_movement['user'])
min_30_sec_users
```

```python
df_first_last = pd.DataFrame(columns={'first', 'last', 'number'})
for user in min_30_seconds_users:
    df_temp = df_min_30_seconds_strong_movement[df_min_30_seconds_strong_movement['user'] == user]
    new_row = {'first':df_temp.iloc[0, 2], 'last':df_temp.iloc[-1, 3], 'number':min_30_sec_users[user]}
    df_first_last = df_first_last.append(new_row, ignore_index=True)
df_first_last
```

```python
# extract from filenames timestamp in the format: 2019-10-01 11:35:00
for i, column in enumerate(['first', 'last']):
    timepoint = df_first_last[column].str.replace('screenshot-2019101-', '2019-10-01 ')
    timepoint = timepoint.str.replace('screenshot-2019107-', '2019-10-07 ')
    timepoint = timepoint.str.replace('screenshot-20191015-', '2019-10-15 ')
    timepoint = timepoint.str.replace('.png', '')
    timepoint = timepoint.str[:-3]
    split = timepoint.str.split(' ')
    timestamp_day, time = zip(*split)

    new_time = []
    new_time_maybe = [None] * len(time)
    for idx, t in enumerate(time):
        #print('len', len(t), t)
        if len(t) == 6:
            new = t[:2] + ':' + t[2:4] + ':' + t[4:]
        elif len(t) == 4:
            new = t[:2] + ':0' + t[2:3] + ':0' + t[3:]
        elif len(t) == 5 and int(t[2:4]) > 60:
            new = t[:2] + ':0' + t[2:3] + ':' + t[3:]
        elif len(t) == 5 and int(t[3:5]) > 60:
            new = t[:2] + ':' + t[2:4] + ':0' + t[4:]
        elif len(t) == 5 and t[3] == '0':
            new = t[:2] + ':' + t[2:4] + ':0' + t[4:]
        # not possible to say for sure what time it is
        else:
            new_time_maybe[idx] = t[:2] + ':' + t[2:4] + ':0' + t[4:]
            new = t[:2] + ':0' + t[2:3] + ':' + t[3:]
        new_time.append(new)

    if i == 0:
        df_first_last['start'] = pd.Series(new_time)
        df_first_last['start_maybe'] = pd.Series(new_time_maybe)
    else:
        df_first_last['end'] = pd.Series(new_time)
        df_first_last['end_maybe'] = pd.Series(new_time_maybe)
        
df_first_last['timestamp_day'] = timestamp_day
df_first_last
```

```python
#start_maybe_stamp = pd.to_datetime((df_first_last['timestamp_day'] + ' ' + df_first_last['start_maybe']))
#print(start_maybe_stamp)

df_first_last['start_stamp'] = pd.to_datetime((df_first_last['timestamp_day'] + ' ' + df_first_last['start']))
df_first_last['end_stamp'] = pd.to_datetime((df_first_last['timestamp_day'] + ' ' + df_first_last['end']))
df_first_last['start_maybe_stamp'] = pd.to_datetime((df_first_last['timestamp_day'] + ' ' + df_first_last['start_maybe']))
df_first_last['end_maybe_stamp'] = pd.to_datetime((df_first_last['timestamp_day'] + ' ' + df_first_last['end_maybe']))

df_first_last['start'] = pd.to_datetime((df_first_last['start']), format='%H:%M:%S').dt.time
df_first_last['end'] = pd.to_datetime((df_first_last['end']), format='%H:%M:%S').dt.time
df_first_last['start_maybe'] = pd.to_datetime((df_first_last['start_maybe']), format='%H:%M:%S').dt.time
df_first_last['end_maybe'] = pd.to_datetime((df_first_last['end_maybe']), format='%H:%M:%S').dt.time

df_first_last.dtypes
```

```python
for idx in [5, 8, 10, 13, 19, 23, 32, 33]:
    end_start = not_sure_times.loc[idx, 'end_stamp'] - not_sure_times.loc[idx, 'start_stamp']
    end_maybe_start = not_sure_times.loc[idx, 'end_maybe_stamp'] - not_sure_times.loc[idx, 'start_stamp']
    end_start_maybe = not_sure_times.loc[idx, 'end_stamp'] - not_sure_times.loc[idx, 'start_maybe_stamp']
    end_maybe_start_maybe = not_sure_times.loc[idx, 'end_maybe_stamp'] - not_sure_times.loc[idx, 'start_maybe_stamp']
    print('idx: {}, number: {}, end-start: {}, end_maybe-start: {}, end_start-maybe: {}, end-maybe_start-maybe: {}'
          .format(idx, not_sure_times.loc[idx, 'number'], end_start, end_maybe_start, end_start_maybe, end_maybe_start_maybe))
```
