### Import libraries

### Load data and add columns with metadata

```python
# read and show the json file with vectors
dir_path = '../../../../../Praktikum_data/tracking_test/'

def set_file_paths(ext):

        # set data path for all files with the given extension
        dir_paths = dir_path + '*' + ext
        file_paths = glob.glob(dir_paths)
        # check if any files are found
        if not file_paths:
            sys.exit("No files found, check the path")
        return file_paths
```

```python
df = pd.DataFrame()
ex_data_rows = 90 #1 minute = 60 * 10 rows
min_or_data_rows = 10000
# part of data used for clustering
split_sec = 30
splits_in_ex_data = int(ex_data_rows / split_sec)

split = 0

# set file paths for text files
file_paths = set_file_paths('csv')
for file_path in file_paths:
    file_name = os.path.splitext(os.path.basename(file_path))[0]
    splitted = file_name.split('-')
    user_data = pd.read_csv(file_path)
    
    data_lenght = len(user_data)
    if data_lenght < ex_data_rows:
        print("did not use data {}, to small dataset: {}!".format(file_name, data_lenght))
    else:
        print("dataset length: {}".format(data_lenght))

        if min_or_data_rows > data_lenght:
            min_or_data_rows = data_lenght
            
        # add columns with metadata: user_id, study_location, study_variant
        user_data.insert(0, 'Location', splitted[1])
        user_data.insert(0, 'Variant', splitted[2])
        user_data.insert(0, 'Session ID', splitted[4])

        # add column for distinction between orientation and exploration (last 1 minute) data
        user_data.insert(0, 'Period', 'orientation')
        user_data.iloc[-ex_data_rows:, 0] = 'exploration'

        # add also columns with seconds strarted from 0 for both orientation and exploration data
        user_data.insert(0, 'Seconds Intern', range(0, data_lenght))
        user_data.iloc[-ex_data_rows:, 0] = range(0, ex_data_rows)
        user_data['Seconds Intern'] = user_data['Seconds Intern'].apply(lambda row: float(row/10.0))

        # data are saved 10 times per second
        user_data.insert(0, 'Seconds', range(0, data_lenght))
        user_data['Seconds'] = user_data['Seconds'].apply(lambda row: float(row/10.0))
        
        # split user data for parts to cluster
        split_count = int(data_lenght/split_sec)
        
        print("split_count", split_count)
        user_data.insert(0, 'Split', 0)
        for i in range(split_count+1):
            #print("i", i)
            if i == 0:
                user_data.iloc[:split_sec, 0] = split
                #print(split)
                split += 1
            else:
                user_data.iloc[(i)*split_sec:(i+1)*split_sec, 0] = split
                #print(split)
                split += 1
        for i in range(splits_in_ex_data):
            user_data.iloc[-(splits_in_ex_data-i)*split_sec:, 0] = split-splits_in_ex_data-i+2
            #print(split-splits_in_ex_data-i+2)

        if df.empty:
            df = user_data
        else:
            df = df.append(user_data)
            
min_or_data_rows -= ex_data_rows
print("length of exploration data: {}".format(ex_data_rows))
print("min length of orientation data: {}".format(min_or_data_rows))
df
```

### Compute and visualize angles differences in all three directions

```python
# calculate angle differences
data_lenght = len(df)
df.insert(0, 'Diff Pitch', range(0, data_lenght))
df.insert(0, 'Diff Yaw', range(0, data_lenght))
df.insert(0, 'Diff Roll', range(0, data_lenght))

df.insert(0, 'Abs Diff Pitch', range(0, data_lenght))
df.insert(0, 'Abs Diff Yaw', range(0, data_lenght))
df.insert(0, 'Abs Diff Roll', range(0, data_lenght))
for i in range(data_lenght):
    if i == 0:
        df.loc[0, 'Diff Pitch'] = 0
        df.loc[0, 'Diff Yaw'] = 0
        df.loc[0, 'Diff Roll'] = 0
        
        df.loc[0, 'Abs Diff Pitch'] = 0
        df.loc[0, 'Abs Diff Yaw'] = 0
        df.loc[0, 'Abs Diff Roll'] = 0
    else:
        # TODO: use an absolute value?
        df.loc[i, 'Diff Pitch'] = float(df.loc[i-1, 'Pitch']) - float(df.loc[i, 'Pitch'])
        df.loc[i, 'Diff Yaw'] = float(df.loc[i-1, 'Yaw']) - float(df.loc[i, 'Yaw'])
        df.loc[i, 'Diff Roll'] = float(df.loc[i-1, 'Roll']) - float(df.loc[i, 'Roll'])
        
        df.loc[i, 'Abs Diff Pitch'] = abs(float(df.loc[i-1, 'Pitch']) - float(df.loc[i, 'Pitch']))
        df.loc[i, 'Abs Diff Yaw'] = abs(float(df.loc[i-1, 'Yaw']) - float(df.loc[i, 'Yaw']))
        df.loc[i, 'Abs Diff Roll'] = abs(float(df.loc[i-1, 'Roll']) - float(df.loc[i, 'Roll']))
```

### Clustering

```python
# TODO: cluster data per 20-30 seconds, group by 'Split'
df_or_cluster = df[['Diff Pitch', 'Diff Yaw', 'Diff Roll']].copy()

# Change slightly zero data because of scipy bug
df_or_cluster['Diff Pitch'] = df_or_cluster['Diff Pitch'].replace(0.000000, 0.000001)
df_or_cluster['Diff Yaw'] = df_or_cluster['Diff Yaw'].replace(0.000000, 0.000002)
```

```python
# The input y may be either a 1d condensed distance matrix or a 2d array of observation vectors.
Z = hac.linkage(df_or_cluster, method='single', metric='correlation')

# Plot dendogram
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
hac.dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
def print_clusters(timeSeries, Z, k):
    # Finds a minimum threshold ``r`` so that the cophenetic distance between any two original
    # observations in the same flat cluster is no more than ``r`` and no more than `t` flat clusters are formed.
    results = fcluster(Z, k, criterion='maxclust')

    s = pd.Series(results)
    clusters = s.unique()

    for c in clusters:
        cluster_indeces = s[s==c].index
        print("Cluster %d number of entries %d" % (c, len(cluster_indeces)))
```
