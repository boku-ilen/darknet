```python
import pandas as pd
from pandas.io.json import json_normalize
from collections import Counter
import plotly.express as px
import math
```

```python
path = '../../Praktikum_data/workshops/vectors/vectors_0.01-191415.json'

json_file = pd.read_json(path)
json_file
```

```python
df_fields = pd.io.json.json_normalize(json_file['fields'])
df_fields
```

```python
# coordinates order: x (pitch), z (roll), y (yaw) 
df_fields['location_SRID'], df_fields['location_POINT Z'] = df_fields['location'].str.split(';', 1).str
df_fields['viewport_SRID'], df_fields['viewport_POINT Z'] = df_fields['viewport'].str.split(';', 1).str
```

```python
df_fields = df_fields.drop(['location'], axis=1)
df_fields = df_fields.drop(['viewport'], axis=1)
```

```python
df_fields
```

```python
dropped, df_fields['location_POINT Z'] = df_fields['location_POINT Z'].str.split('(', 1).str
df_fields['location_POINT Z'], dropped = df_fields['location_POINT Z'].str.split(')', 1).str
df_fields['location_POINT Z_x'], df_fields['location_POINT Z_z'], df_fields['location_POINT Z_y'] = df_fields['location_POINT Z'].str.split(' ', 2).str
df_fields = df_fields.drop(['location_POINT Z'], axis=1)
```

```python
dropped, df_fields['viewport_POINT Z'] = df_fields['viewport_POINT Z'].str.split('(', 1).str
df_fields['viewport_POINT Z'], dropped = df_fields['viewport_POINT Z'].str.split(')', 1).str
df_fields['viewport_POINT Z_x'], df_fields['viewport_POINT Z_z'], df_fields['viewport_POINT Z_y'] = df_fields['viewport_POINT Z'].str.split(' ', 2).str
df_fields = df_fields.drop(['viewport_POINT Z'], axis=1)
```

```python
df_fields['timestamp_day'], df_fields['timestamp_hour'] = df_fields['timestamp'].str.split('T', 1).str
df_fields = df_fields.drop(['timestamp'], axis=1)
```

```python
df_fields
```

```python
df_model = pd.DataFrame.from_dict(json_file['model'])
df_pk = pd.DataFrame.from_dict(json_file['pk'])
df = df_model.join(df_pk).join(df_fields)
df
```

```python
Counter(df['timestamp_day'])
```

```python
# take only 1.10, 7.10 and 15.10
df1 = df[df['timestamp_day'] == '2019-10-01']
df2 = df[df['timestamp_day'] == '2019-10-07']
df3 = df[df['timestamp_day'] == '2019-10-15']
df = df1.append(df2).append(df3)
df
```

```python
# set new ids
df.insert(0, 'ID', range(0, len(df)))
df.set_index('ID', inplace=True)
df
```

```python
# rotation by y-axis: north/west etc. so don't use for clustering directly
df['yaw'] = df.apply(lambda row: math.degrees(math.atan2(float(row['viewport_POINT Z_x']), float(row['viewport_POINT Z_z']))), axis=1)
fig = px.line(df, y="yaw", color="model")
fig.show()
```

```python
# rotation by x-axis: up/down, take for analysis 
df['pitch'] = df.apply(lambda row: math.degrees(math.asin(float(row['viewport_POINT Z_y']))), axis=1)
fig = px.line(df, y="pitch", color="model")
fig.show()
```

```python
# calculate vector lenght to the previous direction point
#df.insert(0, 'diff', range(0, len(df)))
# just testing
for i in range(len(df)):
    if i == 0:
        df.loc[0, 'diff'] = 0
    else:
        diff_x = float(df['viewport_POINT Z_x'][i-1]) - float(df['viewport_POINT Z_x'][i])
        diff_z = float(df['viewport_POINT Z_z'][i-1]) - float(df['viewport_POINT Z_z'][i])
        diff_y = float(df['viewport_POINT Z_y'][i-1]) - float(df['viewport_POINT Z_y'][i])
                                                             
        df.loc[i, 'diff'] = (diff_x**2 + diff_z**2 + diff_y**2) ** (1.0 / 2)
```

```python
df
```

```python
fig = px.line(df, y="diff", color="model")
fig.show()
```
